__[문제1]__

한 주머니에 사탕이 5개씩 들어있습니다.

바리스타인 나에게는 3개의 주머니가 있습니다!

근데 길가다가 사탕을 주웠습니다. 그래서 사탕 3개가 주머니에 더 추가되었습니다.

최종적으로 주머니에 들어있는 사탕의 총 개수는 얼마인가요?

<br>

```python
#사탕의 갯수
a = 5
#내가 가지고있는 주머니의 갯수
b = 3
#길에서 주운 사탕의 갯수
c = 3

print("내가 가지고 있는 사탕의 총 갯수는", (a * b) + c, "개 입니다.")
```

__[문제2]__

10명이 각자 4개의 공을 가지고 있습니다.

운이 좋은 3명이 2개의 공을 더 받았습니다.

10명이 가진 모든 공의 총 개수를 구하세요.


```python
#각자 가지고 있는 공의 갯수
a= 4
#사람의 수
b= 10
#공을 더 받은 사람의 수
c= 3
#받은 공의 수
d= 2

print("가지고 있는 공의 총 갯수는", ((a * b) + (c * d)), "개 입니다.")
```

__[문제3]__

5명의 학생이 3시간씩 공부했습니다.

이상한 강박관념이 있는 나는 모든 시간을 분으로 계산하고 싶습니다.

대신 계산해주세요 !


```python
#사람의 수 
a = 5
#공부한 시간
b = 3
#시간을 분으로 환산
print("총 공부한 시간은", a * b * 60, "분 입니다.")
```

__[문제4]__

차가 9리터의 연료를 가지고 있습니다.

연료가 3일 동안 계속 줄어들며, 하루에 2리터씩 소비됩니다.

3일 뒤 차에 남은 연료는 몇 리터인가요?

```python
#차의 연료
a = 9
#연료가 줄어는 일수
b = 3
#하루에 소비되는 연료
c = 2
print("남은 연료는", a - (b * c), "리터 입니다.")
```

__[문제5]__

어떤 정사각형의 넓이가 144cm²입니다.

정사각형의 한 변의 길이를 구하세요.


```python
#정사각형의 넓이
a = 144
print("정사각형의 한 변의 길이는", a ** 0.5, "미터 입니다.")
#or
print("정사각형의 한 변의 길이는", a ** (1/2), "미터 입니다.")
```

__[문제6]__

오즈 코딩 스쿨 백엔드 과정에 참여한 수강생 중 5명의 수강생이 자율적으로 스터디 그룹을 결성했습니다.

스터디 그룹 내에는 한 주간 배운 내용으로 매주 쪽지 시험을 보는 룰이 있습니다.

이 쪽지 시험의 평균이 70점이 넘으면 주말 스터디는 없습니다.

이들의 평균 점수를 대신 계산해주세요.

<br>
A: 65, B: 75, C: 50, D: 80, E: 60

<br>


```python
a = 65
b = 75
c = 50
d = 80
e = 60

print("평균은", (a + b + c + d + e) / 5, "점 입니다.")
```
